## we check if there is title in attribues,if yes we print it. We just want href and title , the others we dont want 

```python
links = soup.find_all('a')
for link in links:
    if 'title' in link.attrs:
     print (link)
```

## "The set" guaratee the "title" appears only one time 

```python
links = soup.find_all('a')
for link in links:
    S1 = set(['title','href'])
    S2 = set(link.attrs.keys())
    if S1 == S2:
     print (link)
```

## Put all the link into array,then remove the last five.  In the end we will have the adress of every character.  

```python
links = soup.find_all('a')
pages=[]
for link in links:
    S1 = set(['title','href'])
    S2 = set(link.attrs.keys())
    if S1 == S2:
        pages.append((link['href'],link.text))
pages = pages[:-5]
for page in pages:
    print(page)
```

## We split on the "/", and get the last element 

```python
links = soup.find_all('a')
pages =[]
for link in links:
    S1 = set(['title','href'])
    S2 = set(link.attrs.keys())
    if S1 == S2:
        pages.append((link['href'],link['href'].split('/')[-1]))
pages = pages[:-5]
pages = set(pages)  
```

## Each time the list contains 2 things- the link and the name 

```python
for href,name in pages:
    if name == 'Aegon_V_Targaryen':
        print(href)
        r=requests.get("https://awoiaf.westeros.org" + href)
        with open(name+'.html','w+') as fp:
            fp.write(r.text)
        
```

##  We are looking for full name. Tbody-branche of whole doc. 

```python
tbody = soup.find("table", class_="infobox").tbody
rows = tbody.find_all("tr")
for row in rows:
    for child in row.children:
        if child.name == 'th':
            if row.th.text == 'Full Name':
                print(row.td.text)
```
